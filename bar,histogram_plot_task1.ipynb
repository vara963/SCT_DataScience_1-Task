```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set Seaborn theme
sns.set_theme(style="darkgrid")

# ------------------------
# Generate Sample Dataset
# ------------------------
np.random.seed(42)

data = {
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=100),
    'Age': np.random.normal(loc=30, scale=10, size=100).astype(int)
}

df = pd.DataFrame(data)
df['Age'] = df['Age'].clip(lower=0)

# ------------------------
# Bar Chart: Gender Distribution
# ------------------------
plt.figure(figsize=(8, 4))
custom_palette = {'Male': '#1f77b4', 'Female': '#ff69b4', 'Other': '#8c564b'}
ax = sns.countplot(x='Gender', hue='Gender', data=df, palette=custom_palette, legend=False)

plt.title('Distribution of Respondents by Gender')
plt.xlabel('Gender Category')
plt.ylabel('Number of Participants')

# Add value labels
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline')

plt.tight_layout()
plt.show()

# ------------------------
# Histogram: Age Distribution
# ------------------------
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=10, kde=True, color='blue')

plt.title('Distribution of Participant Ages')
plt.xlabel('Age in Years')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
